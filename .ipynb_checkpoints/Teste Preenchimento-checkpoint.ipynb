```python
def fix_city_and_uf(row):
    # Divide a coluna `uf` em partes
    parts = row['UF'].split(' - ')
    if len(parts) == 2:
        # Se houver duas partes, ajusta cidade e UF
        cidade_corrigida = f"{row['Cidade']} {parts[0]}".strip()
        uf_corrigida = parts[1]
    else:
        # Caso esteja correto, mantém os valores
        cidade_corrigida = row['Cidade']
        uf_corrigida = row['UF']
    
    return pd.Series([cidade_corrigida, uf_corrigida])
```

```python
teste = df.apply(fix_city_and_uf, axis=1)
```

```python
teste
```

```python
nomes_uf = teste.apply(lambda row: f"{row[0].title()} - {row[1].upper()}", axis=1)
```
